Let's calculate how much likely are people to have had hypertension or heart disease if they suffered a stroke compared to if they hadn't, by calculating the percentages within the group

```python
stroke_nostroke_df = pd.DataFrame(columns=['stroke', 'no stroke'])
stroke_nostroke_df

#hypertension
stroke_hypertension = stroke_df[['hypertension']].value_counts()
nostroke_hypertension = nostroke_df[['hypertension']].value_counts()
new_row = {
    'stroke': stroke_hypertension[1] * 100 / stroke_hypertension[0],
    'no stroke': nostroke_hypertension[1] * 100 / nostroke_hypertension[0]
}
new_df = pd.DataFrame([new_row])
new_df.index = ['hypertension']

stroke_nostroke_df = pd.concat([stroke_nostroke_df, new_df])

# heart disease
stroke_heart = stroke_df[['heart_disease']].value_counts()
nostroke_heart = nostroke_df[['heart_disease']].value_counts()

new_row = {
    'stroke': stroke_heart[1] * 100 / stroke_heart[0],
    'no stroke': nostroke_heart[1] * 100 / nostroke_heart[0]
}

new_df = pd.DataFrame([new_row])
new_df.index = ['heart_disease']

stroke_nostroke_df = pd.concat([stroke_nostroke_df, new_df])

# smoking status
stroke_smokes = stroke_df[['smoking_status']].value_counts()
nostroke_smokes = nostroke_df[['smoking_status']].value_counts()

new_row = {
    'stroke': (stroke_smokes[1] + stroke_smokes[3])  * 100 / (stroke_smokes[0] + stroke_smokes[2]),
    'no stroke': (nostroke_smokes[1] + nostroke_smokes[3])  * 100 / (nostroke_smokes[0] + nostroke_smokes[2]),
}

new_df = pd.DataFrame([new_row])
new_df.index = ['smokes']

stroke_nostroke_df = pd.concat([stroke_nostroke_df, new_df])

stroke_nostroke_df['proportion'] = stroke_nostroke_df['stroke'] / stroke_nostroke_df['no stroke']
stroke_nostroke_df
```

It would seem that patients who suffer a stroke are about 3.7 times as likely to have hypertension and about 4.7 times as likely to have had problems with heart disease. Interestingly, in this dataset the percentage of smoking patients doesnt seem to be related to the chance of suffering a stroke, even though, it is often thought of as a possible important factor

Looking at percentage comparisons:
* Almost all of the patients who suffered a stroke were over the age of 40
* The above calculations related to hypertension and heart disease
* Most who have suffered a stroke are outside the typically cited "healthy" BMI range
* There's a noticeable, but not as significant relation to having been married and suffering a stroke

### Moving the classification threshold
By default, binary classifiers in sklearn, regardless of class distribution, will set the classification threshold at 0.5, meaning, assuming that the predictions should more or less be split evenly between the classes, if such a balance exists in the dataset. As seen, our dataset does not fit such criterion. Especially when looking at the precision-recall curve, we can see, the chance model operating on a much lower/higher threshold and our curve following suit. Therefore, let's try moving the prediction threshold closer to one of the classes, hopefully achieving better results for the minority class.

```python
preds_thresh = forest.predict_proba(X_test)
preds_thresh
```

Looking at the probabilities for each class, indeed we can see, the model is typically very sure of the predictions belonging to class 0 (majority), therefore, the default threshold of 0.5 is not having any effect on the results

```python
preds_thresh[:,0].mean()
```

```python
preds_thresh[:,1].mean()
```

Interestingly, when looking at the means of the prediction probabilities, we get a proportion very similar to the proportion of the classes in our data.

```python
preds_thresh = (forest.predict_proba(X_test)[:,1] >= 0.1).astype('int')
preds_thresh
```

```python
unique, counts = np.unique(preds_thresh, return_counts=True)
print(np.asarray((unique, counts)).T)
```

It seems we might have gone overboard with the threshold, still, let's examine some scoring reports

As stated in the preamble, we might indeed be most interested in the maximization of the recall score for positive stroke predictions, still, we definitely should consider establishing a stopping point, as we don't want to falsely predict a positive chance of suffering a stroke, as that, in our imaginary scenario might for example cause too many further checkups, reducing the care given to actual patients at risk of suffering a stroke.

## Class weight explorations
Using scikit-learn, for some alogrithms, a parameter of class weights is available. Setting this parameter is supposed to guide the model training cost functions to assign higher importance to the minority class, by setting it's weight higher, allowing the model to pay more attention to patterns related to this class and reducing the bias towards the majority class.

Let's start with a simple class weights formula, stated as: weight_i = n_samples / (n_classes * n_samples_i)

```python
stroke_weight = df_enc.shape[0] / (2 * df_enc[df_enc['stroke'] == 1].shape[0])
nostroke_weight = df_enc.shape[0] / (2 * df_enc[df_enc['stroke'] == 0].shape[0])
print(f'stroke weight: {stroke_weight}, no stroke: {nostroke_weight}')
```

#### Let's start without scaling

```python
X = df_enc.drop(['stroke'], axis=1)
y = df_enc['stroke']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)

forest_weighted = RandomForestClassifier(random_state=1, class_weight = {0: nostroke_weight, 1: stroke_weight})
# forest_weighted = RandomForestClassifier(random_state=1, class_weight = "balanced")
forest_weighted.fit(X_train, y_train)

preds_weighted = forest.predict(X_test)
acc_weighted = accuracy_score(y_test, preds)
print('accuracy: ', acc_weighted)

precision_weighted, recall_weighted, _ = precision_recall_curve(y_test, preds_weighted)
print(f'precision: {precision_weighted}, recall: {recall_weighted}')

display_weighted = PrecisionRecallDisplay.from_estimator(
    forest_weighted, X_test, y_test, name="RandomForest", plot_chance_level = True, pos_label=1
)
# _ = display.ax_.set_title("2-class Precision-Recall curve")
# display.plot()
```

```python
forest_weighted_report = classification_report(y_test, preds_weighted, target_names=['no stroke', 'stroke'])
print(forest_weighted_report)
```

No improvement so far, let's experiment with changing the prediction threshold.

An improvement! Not a huge one, an improvement nonetheless, with both more 'stroke' occurences predicted correctly and a lower number of false positives.

### Finetuning the random forests model - grid search
perhaps important to mention - so far we mostly used a random train/test split, with stratify=True, which ensures an even proportion of classes in both subsets - very important to achieve accurate testing results, especially with grid searching and cross-validation in mind. A very similar behaviour can be achieved using 'StratifiedKFold' from the same library

```python
param_grid = {
    'min_samples_split': [3, 5, 10], 
    'n_estimators' : [100, 300],
    'max_depth': [3, 5, 15, 25],
    'max_features': [3, 5, 10, 20]
}

scorers = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score)
}
```

```python
def grid_search_wrapper(clf, refit_score='recall_score'):
    """
    fits a GridSearchCV classifier using refit_score for optimization
    prints classifier performance metrics
    """
    
    skf = StratifiedKFold(n_splits=10)
    grid_search = GridSearchCV(clf, param_grid, scoring=scorers, refit=refit_score,
                           cv=skf, return_train_score=True, n_jobs=-1)
    grid_search.fit(X_train.values, y_train.values)

    # make the predictions
    y_pred = grid_search.predict(X_test.values)

    print('Best params for {}'.format(refit_score))
    print(grid_search.best_params_)

    # confusion matrix on the test data.
    print('\nConfusion matrix of Random Forest optimized for {} on the test data:'.format(refit_score))
    print(pd.DataFrame(confusion_matrix(y_test, y_pred),
                 columns=['pred_neg', 'pred_pos'], index=['neg', 'pos']))
    return grid_search
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=1)
clf = RandomForestClassifier(n_jobs=-1, random_state=1, class_weight={0: nostroke_weight, 1: stroke_weight})

grid_search_clf = grid_search_wrapper(clf, refit_score='recall_score')
```

```python
results = pd.DataFrame(grid_search_clf.cv_results_)
results = results.sort_values(by='mean_test_recall_score', ascending=False)
results[['mean_test_precision_score', 'mean_test_recall_score', 'mean_test_accuracy_score', 'param_max_depth', 'param_max_features', 'param_min_samples_split', 'param_n_estimators']].round(3).head()
```

### From: https://towardsdatascience.com/fine-tuning-a-classifier-in-scikit-learn-66e048c21e65

```python
y_scores = grid_search_clf.predict_proba(X_test)[:, 1]
```

```python
p, r, thresholds = precision_recall_curve(y_test, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    """
    Modified from:
    Hands-On Machine learning with Scikit-Learn
    and TensorFlow; p.89
    """
    plt.figure(figsize=(8, 8))
    plt.title("Precision and Recall Scores as a function of the decision threshold")
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.ylabel("Score")
    plt.xlabel("Decision Threshold")
    plt.legend(loc='best')
```

```python
# use the same p, r, thresholds that were previously calculated
plot_precision_recall_vs_threshold(p, r, thresholds)
```

```python
best_model = grid_search_clf.best_estimator_
best_model
```
